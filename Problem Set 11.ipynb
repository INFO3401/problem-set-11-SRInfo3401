```python
#Top section In Class Pratice. 
#BELOW section PROBLEM SET 11.
import pandas as pd
import altair as alt
alt.renderers.enable('notebook')

#Can replace with any dataframe
from vega_datasets import data
cars = data.cars()

print(cars.columns)
```

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin'
).interactive()
```

```python
alt.Chart(cars).mark_bar().encode(
    x='Name',
    y='Cylinders',
)
#.interactive()
```

```python
alt.Chart(cars).mark_bar().encode(
    x='Year',
    y='average(Cylinders)'
)
```

```python
alt.Chart(cars).mark_bar().encode(
    alt.X('Horsepower', bin=True),
    y='count()'
)
```

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Acceleration',
    color='Origin',
    size='Year'
).interactive()
```
